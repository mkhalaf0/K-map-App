```python
#Majd Khalaf
#Honors program
#2-Variable kmap function
import numpy as np


#insert a decimal function and convert it into an array with binary numbers

deci_fun=[]    #Declaring the decimal array, bit array
bit_array=[]
i=0
n=0
c=0
n_var=["A","B"]        #an array for how many variables we have
length=len(n_var)      #to find the number of variables we have
n_bit=2**length       #to find the number of bits we have in total


A3,A2,A1,A0=input("Input Your function in the form of f(A,B)= A3,A2,A1,A0 ").split(",")
element=[A3,A2,A1,A0]
for i in range(n_bit):     #a for loop to input the decimal numbers (SOP) 
         

    if element[i].isdigit():   #if statement to not include any nondigit entry in the array
        int(element[i])      #Convert the number entred from the form of a string to the integer form
        if int(element[i])<n_bit:   #if statement to not include any numbers higher than the number of bits
             deci_fun.insert(i,int(element[i]) )  #insert the elements in the array
deci_fun_noduplct=[]

for i in deci_fun:                       #ensure no duplicate numbers in the decimal array
    if i not in deci_fun_noduplct:       #this line adds the elements that are not in the new array only one time
        deci_fun_noduplct.append(i)

fun_length=len(deci_fun_noduplct)  #find the length of the decimal array


for n in range(fun_length):  # for loop to convert the decimal into binary and put the data in the bit_array
    num_pref=format(deci_fun_noduplct[n], f'#0{length+2}b')  #convert the decimal into binary based on how many bits we need
    num=num_pref[2:]        #delete the part nb' from the binary number
    num_split=num[0]+" "+num[1]  #split the numbers into two and add a space between them /note:only works for 2-Variables
    bit_array.insert(n,num_split) #insert the binary numbers to the bit_array

while len(bit_array)<n_bit:  #a while loop in case the length of the array is less than the need one
    bit_array.append("8 8")   #add (8 8) as a representation of an empty element
    


#Use the binary array to find the K-map

#bit_array=["0 0","0 1","1 0","8 8"]

bit_array_string = [i for j in bit_array for i in j.split()]  #split the array and double the size /depends on the n-v
bit_array_int=[int(i) for i in bit_array_string]  #convert the array into an integer
Shape_2_Variable=int(len(bit_array_int)/2)    #Only for 2-Variable #find the length of the array
fbit_array = np.array(bit_array_int).reshape(Shape_2_Variable,2)  #Only for 2-Variable #reshape the array into a 2d array


row=len(fbit_array[0])      #Only for 2-Variable #Can depend on the number of variables when you join the code together
column=len(fbit_array[0])   #Only for 2-Variable #Can depend on the number of variables when you join the code together

kmap_table=np.array([[0 for row in range(row)] for column in range(column)]) #declare the 2d array for the kmap
for i in range(column):   #start pluging the numbers in the kmap based on the bit array
    for j in range(row):
        if i==fbit_array[0][0] and j==fbit_array[0][1]:     #Only for 2-Variable
            kmap_table[i][j]=1
        elif i==fbit_array[1][0] and j==fbit_array[1][1]:
            kmap_table[i][j]=1
        
        elif i==fbit_array[2][0] and j==fbit_array[2][1]:
             kmap_table[i][j]=1
       
        elif i==fbit_array[3][0] and j==fbit_array[3][1]:
            kmap_table[i][j]=1
                
              
        
                                   
    
        
print(kmap_table)



#Convert a kmap Table to a boolean expression


    #2 VARIABLE K-map
first_term=""
second_term=""
if kmap_table[0][1]==1 & kmap_table[1][1]==1:
    first_term="B"

if kmap_table[0][0]==1 & kmap_table[0][1]==1:
    first_term='Ā'
    
elif kmap_table[0][0]==1:
    first_term="ĀB̅"
    
elif kmap_table[0][1]==1:
    first_term="ĀB"
    
if kmap_table[1][0]==1:
    first_term=="B"

elif kmap_table[1][1]==1:
    first_term==""

if kmap_table[1][0]==1 & kmap_table[1][1]==1:
     first_term="A"
if kmap_table[0][0]==1 & kmap_table[1][0]==1:
    second_term="+B̅"
if kmap_table[0][1]==1 & kmap_table[1][1]==1 and first_term !="B":
    second_term="+B"
if kmap_table[1][0]==1 & kmap_table[0][1]==1:
    first_term ="AB̅"
    second_term="+ĀB"
    
        
if kmap_table.all()==1:
    first_term=" "
    second_term=" "
    
boolean=first_term+second_term
print(boolean)


# case:[[0,1],[0,1]] remember to implement it
        
```
